# Transform

```python
df['budget']= (df['budget']/1000000).round(2)
df['revenue']= (df['revenue']/1000000).round(2)
df
```

```python
def calc_exp(x):
    x = x - x.mean();
    return x;
```

```python
df.groupby('director_name')['budget'].transform(calc_exp)
```

```python
def calc_risky(x):
    x["risky"]= x["budget"] - x['revenue'].mean() >= 0;
    return x
```

```python
df.groupby("director_name").apply(calc_risky)
```
