```python
df1 = pd.DataFrame({'name':["Ram","Shyam","Mukesh","Suresh"],'age':[10,18,60,50]})
df1
```

```python
def calc_age(x):
    if (np.any(x['age'] < 18)):
        x['age'] = 'kid';
    elif (np.any(x['age'] >= 18) & np.any(x['age'] <= 50)):
        x['age'] = 'adult'
    elif (np.any(x['age'] > 50)):
        x['age'] = 'senior';
    return x;

df1.groupby('name').apply(calc_age)
```

```python
temp_bins = [1, 17, 51, 100];
temp_labels = ['kid', 'adult', 'senior']
pd.cut(df1['age'],
      bins=temp_bins,
      labels=temp_labels)
```
