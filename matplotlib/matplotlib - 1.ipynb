```python
xpoints = np.arange(-3,3,0.1)
ypoints = xpoints **2

plt.plot(xpoints, ypoints)
plt.xticks(np.arange(-3,3,0.5), rotation=90)
plt.title('y=x**2')
plt.xlabel('x_labels')
plt.ylabel('y_labels')
plt.show()
```

```python
x = np.arange(-3,3,0.1)
y = x **2
y1 = x**3
y2 = x**4

plt.plot(x, y, label="squares")
plt.plot(x, y1, label="cubes")
plt.plot(x, y2, label="fours")
plt.legend()
plt.show()
```

```python
x = np.arange(-3,3,0.1)
y = x **2
y1 = x**3
y2 = x**4

plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(["squares", "cubes", "fours"])
plt.show()
```

```python
x = np.arange(-3,3,0.1)
y = x **2
y1 = x**3
y2 = x**4

plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)
plt.text(1,1, "(1,1)")
plt.grid() # show the grid
plt.legend(["squares", "cubes", "fours"], loc="upper center", frameon=False) # labels, location, remove frame from legend
plt.show()
```

```python
fruits = [1,2,3]
prices= [10,60,40]

plt.bar(fruits, prices, width=0.2, color='g')
plt.legend(['Fruits'])
plt.show()
```

```python
fruits = [1,2,3]
price1= [10,60,40]
price2= [8,50,35]
plt.bar(fruits, price1, width=0.2)
plt.bar(fruits, price2, width=0.2)
plt.legend(['Market 1', 'Market 2'])
plt.show()
```

```python
fruits = np.array([1,2,3])
price1= np.array([10,60,40])
price2= np.array([8,50,35])
plt.bar(fruits, price1, width=0.2)
plt.bar(fruits + 0.2, price2, width=0.2)
plt.legend(['Market 1', 'Market 2'])
plt.show()
```
